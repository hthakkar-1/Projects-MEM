```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Load data
data = pd.read_csv("/content/odi_Matches_Data.csv")
# Handle missing values
data.dropna(inplace=True)



# Normalize numerical features
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
data_scaled = scaler.fit_transform(data[["Team1 Runs Scored", "Team1 Wickets Fell", "Team1 Extras Rec"]])
data[["Team1 Runs Scored", "Team1 Wickets Fell", "Team1 Extras Rec"]] = data_scaled
features = ["Team1 Runs Scored", "Team1 Wickets Fell", "Team1 Extras Rec"]
target = "high_score"
data[target] = (data["Team1 Runs Scored"] > 300).astype(int)
X_train, X_test, y_train, y_test = train_test_split(data[features], data[target], test_size=0.2)

model = DecisionTreeClassifier(max_depth=5)
model.fit(X_train, y_train)
# Evaluation
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Model accuracy:", accuracy)

# Feature importance
feature_importance = model.feature_importances_
sorted_features = sorted(zip(feature_importance, features), reverse=True)

# Create a bar chart to visualize feature importance
plt.figure(figsize=(10, 6))
plt.barh([f[1] for f in sorted_features], [f[0] for f in sorted_features])
plt.xlabel("Feature Importance")
plt.ylabel("Feature")
plt.title("Decision Tree Feature Importance")
plt.show()
# Use Plotly to create interactive dashboards
import plotly.express as px

# Explore batting score distribution
fig = px.histogram(data, x="Team1 Runs Scored", color="high_score")
fig.show()

# Compare batting and bowling scores for high and low scoring matches
fig = px.scatter(data, x="Team1 Runs Scored", y="Team1 Wickets Fell", color="high_score")
fig.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from matplotlib import pyplot as plt

# Load datasets
t20_matches = pd.read_csv("t20_matches.csv")
odi_matches_data = pd.read_csv("odi_Matches_Data.csv")

# Define features and target variables
features_t20 = t20_matches[["innings1_runs", "innings2_runs", "innings1_wickets", "innings2_wickets"]]
target_t20 = t20_matches["innings1_runs >= 180","innings2_runs >= 180"]

features_odi = odi_matches_data[["Team1 Runs Scored", "Team2 Runs Scored", "Team1 Wickets Fell", "Team2 Wickets Fell"]]
target_odi = odi_matches_data["Team1 Runs Scored >= 300","Team2 Runs Scored >= 300"]

# Split data into training and testing sets
X_train_t20, X_test_t20, y_train_t20, y_test_t20 = train_test_split(features_t20, target_t20, test_size=0.2)
X_train_odi, X_test_odi, y_train_odi, y_test_odi = train_test_split(features_odi, target_odi, test_size=0.2)

# Define and train models
models = {
    "Linear Regression": LinearRegression(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "SVM": SVC(),
    "Logistic Regression": LogisticRegression(),
    "Ridge": Ridge(),
}


for name, model in models.items():
    model.fit(X_train_t20, y_train_t20)
    scores_t20 = cross_val_score(model, X_test_t20, y_test_t20, cv=5)
    print(f"{name}: T20 Cross-Validation Accuracy: {scores_t20.mean():.4f}")

for name, model in models.items():
    model.fit(X_train_odi, y_train_odi)
    scores_odi = cross_val_score(model, X_test_odi, y_test_odi, cv=5)
    print(f"{name}: ODI Cross-Validation Accuracy: {scores_odi.mean():.4f}")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression, Ridge
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Load your datasets (replace 'your_path' with the actual path or URL)
t20_matches = pd.read_csv('t20_matches.csv')
odi_matches_data = pd.read_csv('odi_Matches_Data.csv')

# Adjust the columns according to your actual data

# For T20Is
t20_features = t20_matches[["innings1_runs", "innings2_runs", "innings1_wickets", "innings2_wickets"]]
t20_labels = t20_matches['win_by_runs','win_by_wickets']

# For ODIs
odi_features = odi_matches_data[["Team1 Runs Scored", "Team2 Runs Scored", "Team1 Wickets Fell", "Team2 Wickets Fell"]]
odi_labels = odi_matches_data['Team 1 Runs Scored >= 300','Team 2 Runs Scored >= 300 ']

# Split the data into training and testing sets
t20_features_train, t20_features_test, t20_labels_train, t20_labels_test = train_test_split(
    t20_features, t20_labels, test_size=0.2, random_state=42
)

odi_features_train, odi_features_test, odi_labels_train, odi_labels_test = train_test_split(
    odi_features, odi_labels, test_size=0.2, random_state=42
)

# Define models
linear_regression = LinearRegression()
decision_tree = DecisionTreeClassifier()
random_forest = RandomForestClassifier()
svm_classifier = SVC()
logistic_regression = LogisticRegression()
ridge_classifier = Ridge()

# Train models
linear_regression.fit(t20_features_train, t20_labels_train)
decision_tree.fit(t20_features_train, t20_labels_train)
random_forest.fit(t20_features_train, t20_labels_train)
svm_classifier.fit(t20_features_train, t20_labels_train)
logistic_regression.fit(t20_features_train, t20_labels_train)
ridge_classifier.fit(t20_features_train, t20_labels_train)

# Predictions
linear_regression_predictions = linear_regression.predict(t20_features_test)
decision_tree_predictions = decision_tree.predict(t20_features_test)
random_forest_predictions = random_forest.predict(t20_features_test)
svm_predictions = svm_classifier.predict(t20_features_test)
logistic_regression_predictions = logistic_regression.predict(t20_features_test)
ridge_predictions = ridge_classifier.predict(t20_features_test)

# Evaluate and compare accuracy
linear_regression_accuracy = accuracy_score(t20_labels_test, linear_regression_predictions)
decision_tree_accuracy = accuracy_score(t20_labels_test, decision_tree_predictions)
random_forest_accuracy = accuracy_score(t20_labels_test, random_forest_predictions)
svm_accuracy = accuracy_score(t20_labels_test, svm_predictions)
logistic_regression_accuracy = accuracy_score(t20_labels_test, logistic_regression_predictions)
ridge_accuracy = accuracy_score(t20_labels_test, ridge_predictions)

# Print accuracies
print(f'Linear Regression Accuracy: {linear_regression_accuracy}')
print(f'Decision Tree Accuracy: {decision_tree_accuracy}')
print(f'Random Forest Accuracy: {random_forest_accuracy}')
print(f'SVM Accuracy: {svm_accuracy}')
print(f'Logistic Regression Accuracy: {logistic_regression_accuracy}')
print(f'Ridge Accuracy: {ridge_accuracy}')
```
