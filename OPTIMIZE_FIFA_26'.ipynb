```python
#print("\nWorld Cup Matches Dataset Info:")
#print(Fifa_world_cup22_matches.info())


# Load the datasets
countries = pd.read_csv("World_cup_2018_country.csv")
goals = pd.read_csv("World_cup_2018_goals.csv")
matches = pd.read_csv("World_cup_2018_matches.csv")

# Clean and manipulate datasets

# 1. **Countries Data Cleaning**
print("Cleaning Countries Data...")
# Convert 'Winner' column to a boolean for easier analysis
countries['Winner'] = countries['Winner'].map({'Y': True, 'N': False})
# Check for duplicate rows
countries = countries.drop_duplicates()
# Ensure rankings are integers
countries[['World_ranking', 'Tournament_ranking']] = countries[['World_ranking', 'Tournament_ranking']].astype(int)

# 2. **Goals Data Cleaning**
print("Cleaning Goals Data...")
# Convert 'Date' to datetime format
goals['Date'] = pd.to_datetime(goals['Date'], format='%Y-%m-%d')
# Remove any duplicates
goals = goals.drop_duplicates()
# Clean 'Own_goal' and 'Penalty' columns
goals['Own_goal'] = goals['Own_goal'].map({'Y': True, 'N': False})
goals['Penalty'] = goals['Penalty'].map({'Y': True, 'N': False})

# 3. **Matches Data Cleaning**
print("Cleaning Matches Data...")
# Convert 'Date' to datetime format
matches['Date'] = pd.to_datetime(matches['Date'], format='%Y-%m-%d')
# Fill NaN values in extra time columns with 0
matches[['Home_extra_goals', 'Away_extra_goals']] = matches[['Home_extra_goals', 'Away_extra_goals']].fillna(0)
# Fill NaN in penalty columns with 0
penalty_cols = ['Home_PK_taken', 'Home_PK_made', 'Away_PK_taken', 'Away_PK_made']
matches[penalty_cols] = matches[penalty_cols].fillna(0)
# Convert appropriate columns to integer
int_columns = ['Home_goals', 'Away_goals', 'Home_shots', 'Away_shots']
matches[int_columns] = matches[int_columns].astype(int)

# Data Manipulation: Merge datasets for visualization preparation
print("Merging Datasets...")
# Merge goals with matches on 'Date' and 'Home' team
goals_matches = pd.merge(goals, matches, how='inner', on=['Date', 'Home', 'Away'])
# Merge with countries to add country-specific info
final_data = pd.merge(goals_matches, countries, how='left', left_on='Home', right_on='Country')

# Save cleaned data for visualization
final_data.to_csv("cleaned_data.csv", index=False)
print("Cleaned and merged data saved to 'cleaned_data.csv'.")
```

```python
# Calculate total goals scored by each team
team_goals = (
    fifa_matches_cleaned.groupby("team1")["number_of_goals_team1"].sum() +
    fifa_matches_cleaned.groupby("team2")["number_of_goals_team2"].sum()
).sort_values(ascending=False).head(10)

# Plot the data
plt.figure(figsize=(10, 6))
team_goals.plot(kind="bar", color="skyblue", edgecolor="black")
plt.title("Top 10 Teams Based on Goals Scored (2018 & 2022)", fontsize=16)
plt.ylabel("Total Goals", fontsize=14)
plt.xlabel("Teams", fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Convert possession columns from strings (e.g., "64%") to numeric
fifa_matches_cleaned["possession_team1"] = (
    fifa_matches_cleaned["possession_team1"].str.replace("%", "").astype(float)
)
fifa_matches_cleaned["possession_team2"] = (
    fifa_matches_cleaned["possession_team2"].str.replace("%", "").astype(float)
)

# Recalculate average possession for each team
possession_team1 = fifa_matches_cleaned.groupby("team1")["possession_team1"].mean()
possession_team2 = fifa_matches_cleaned.groupby("team2")["possession_team2"].mean()

# Combine data for both teams
possession_data = pd.DataFrame({
    "team": possession_team1.index.union(possession_team2.index),
    "average_possession": pd.concat([possession_team1, possession_team2]).groupby(level=0).mean()
}).sort_values(by="average_possession", ascending=False)

# Select top 10 teams by possession for visualization
top_possession_teams = possession_data.head(10)

# Heatmap for possession
import seaborn as sns
plt.figure(figsize=(8, 6))
sns.heatmap(
    top_possession_teams.set_index("team").T,
    annot=True, fmt=".1f", cmap="coolwarm", linewidths=0.5
)
plt.title("Top 10 Teams by Average Possession (%)", fontsize=16)
plt.xlabel("Teams", fontsize=14)
plt.ylabel("Metric", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Read the data
df = pd.read_csv('Fifa_world_cup22_matches.csv')

# Convert possession percentages to numeric
df['possession team1'] = df['possession team1'].str.rstrip('%').astype(float)
df['possession team2'] = df['possession team2'].str.rstrip('%').astype(float)

# 1. Goals and Attempts Comparison
plt.figure(figsize=(15, 8))
metrics = ['number of goals team1', 'number of goals team2',
          'on target attempts team1', 'on target attempts team2']
avg_metrics = df[metrics].mean()

sns.barplot(x=metrics, y=avg_metrics, palette='viridis')
plt.title('Goals and On-Target Attempts Comparison')
plt.xticks(rotation=45)
plt.ylabel('Average Count')
plt.tight_layout()
plt.show()

# 2. Defensive Pressures vs Goals Scored
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df,
                x='defensive pressures applied team1',
                y='number of goals team1',
                size='on target attempts team1',
                sizes=(50, 400),
                alpha=0.6)
plt.title('Defensive Pressures vs Goals Scored (Team 1)')
plt.xlabel('Defensive Pressures')
plt.ylabel('Goals Scored')
plt.tight_layout()
plt.show()

# 3. Passing Efficiency Analysis
plt.figure(figsize=(12, 8))
pass_data = pd.DataFrame({
    'Team': ['Team 1', 'Team 2'],
    'Completed Passes': [df['passes completed team1'].mean(), df['passes completed team2'].mean()],
    'Total Passes': [df['passes team1'].mean(), df['passes team2'].mean()]
})

pass_data_melted = pd.melt(pass_data, id_vars=['Team'], var_name='Pass Type', value_name='Average Passes')
sns.barplot(x='Team', y='Average Passes', hue='Pass Type', data=pass_data_melted, palette='Set2')
plt.title('Passing Analysis by Team')
plt.tight_layout()
plt.show()

# 4. Possession vs Defensive Pressure Relationship
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df,
                x='possession team1',
                y='defensive pressures applied team1',
                size='passes completed team1',
                sizes=(50, 400),
                alpha=0.6)
plt.title('Possession vs Defensive Pressure (Team 1)')
plt.xlabel('Possession (%)')
plt.ylabel('Defensive Pressures')
plt.tight_layout()
plt.show()

# 5. Overall Performance Metrics Distribution
plt.figure(figsize=(15, 10))
performance_metrics = ['number of goals team1', 'on target attempts team1',
                      'passes completed team1', 'defensive pressures applied team1']
df[performance_metrics].boxplot(figsize=(12, 6))
plt.title('Distribution of Key Performance Metrics (Team 1)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Calculate pass completion rates
df['pass_completion_rate_team1'] = (df['passes completed team1'] / df['passes team1']) * 100
df['pass_completion_rate_team2'] = (df['passes completed team2'] / df['passes team2']) * 100

# 6. Pass Completion Rate vs Goals
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df,
                x='pass_completion_rate_team1',
                y='number of goals team1',
                size='total attempts team1',
                sizes=(50, 400),
                alpha=0.6)
plt.title('Pass Completion Rate vs Goals Scored (Team 1)')
plt.xlabel('Pass Completion Rate (%)')
plt.ylabel('Goals Scored')
plt.tight_layout()
plt.show()

# 7. Defensive vs Offensive Balance
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df,
                x='defensive pressures applied team1',
                y='total attempts team1',
                size='number of goals team1',
                sizes=(50, 400),
                alpha=0.6)
plt.title('Defensive vs Offensive Balance (Team 1)')
plt.xlabel('Defensive Pressures')
plt.ylabel('Total Attempts')
plt.tight_layout()
plt.show()

# 8. Performance Metrics Correlation Heatmap
performance_cols = ['number of goals team1', 'on target attempts team1',
                   'passes completed team1', 'defensive pressures applied team1',
                   'possession team1', 'pass_completion_rate_team1']
correlation_matrix = df[performance_cols].corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Performance Metrics Correlation Heatmap')
plt.tight_layout()
plt.show()

# 9. Possession Impact on Different Metrics
fig, axes = plt.subplots(2, 2, figsize=(15, 15))
fig.suptitle('Possession Impact on Different Performance Metrics')

sns.scatterplot(data=df, x='possession team1', y='number of goals team1', ax=axes[0,0])
axes[0,0].set_title('Impact on Goals')

sns.scatterplot(data=df, x='possession team1', y='on target attempts team1', ax=axes[0,1])
axes[0,1].set_title('Impact on On-Target Attempts')

sns.scatterplot(data=df, x='possession team1', y='passes completed team1', ax=axes[1,0])
axes[1,0].set_title('Impact on Completed Passes')

sns.scatterplot(data=df, x='possession team1', y='defensive pressures applied team1', ax=axes[1,1])
axes[1,1].set_title('Impact on Defensive Pressures')

plt.tight_layout()
plt.show()

# 10. Performance Metrics Distribution by Match
plt.figure(figsize=(15, 8))
df_melt = df.melt(value_vars=['number of goals team1', 'on target attempts team1',
                              'defensive pressures applied team1'])
sns.violinplot(x='variable', y='value', data=df_melt)
plt.title('Distribution of Performance Metrics Across Matches')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Assuming 'data' is the FIFA 2022 World Cup match data DataFrame
fifa_22 = data  # Assign the 'data' DataFrame to 'fifa_22'

# Select features and target variable
features = fifa_22[[
    'possession team1', 'possession team2',
    'number of goals team1', 'number of goals team2',
    'defensive pressures applied team1', 'defensive pressures applied team2'
]]

# Convert possession percentages to numeric
features['possession team1'] = features['possession team1'].str.rstrip('%').astype('float')
features['possession team2'] = features['possession team2'].str.rstrip('%').astype('float')

# Target variable: 1 if team1 wins, 0 if team2 wins
fifa_22['outcome'] = (fifa_22['number of goals team1'] > fifa_22['number of goals team2']).astype(int)
target = fifa_22['outcome']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Train a Random Forest Classifier
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Predict outcomes on the test set
predictions = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, predictions)
report = classification_report(y_test, predictions)

print(f"Accuracy: {accuracy}")
print(f"Classification Report:\n{report}")

# Confusion Matrix
cm = confusion_matrix(y_test, predictions)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False)
plt.title("Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()

# Feature Importance
feature_importance = pd.Series(model.feature_importances_, index=features.columns)
feature_importance.nlargest(10).plot(kind="barh")
plt.title("Top 10 Important Features")
plt.xlabel("Feature Importance Score")
plt.show()
```
